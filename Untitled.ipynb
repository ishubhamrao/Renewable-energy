```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('renew.csv')

# Convert 'Year' column to datetime format
data['Year'] = pd.to_datetime(data['Year'], format='%Y')

# Group data by year and calculate mean renewable energy usage
mean_renewable_usage = data.groupby('Year')['Renewables (% equivalent primary energy)'].mean()

# Plot renewable energy adoption trends over time
plt.figure(figsize=(10, 6))
mean_renewable_usage.plot(marker='o', linestyle='-')
plt.title('Renewable Energy Adoption Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Mean Renewables (%)')
plt.grid(True)
plt.show()

# Visualize changes in renewable energy usage across countries
plt.figure(figsize=(12, 8))
sns.boxplot(x='Entity', y='Renewables (% equivalent primary energy)', data=data)
plt.title('Renewable Energy Usage Across Countries')
plt.xlabel('Country')
plt.ylabel('Renewables (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('C:/Users/Neela/bank_data/renew.csv')

# Market Size Analysis
# Calculate total renewable energy production/consumption over the years
total_renewable_energy = data.groupby('Year')['Renewables (% equivalent primary energy)'].sum()

# Plot market size (total renewable energy) over time
plt.figure(figsize=(10, 6))
total_renewable_energy.plot(marker='o', linestyle='-')
plt.title('Renewable Energy Market Size Over Time')
plt.xlabel('Year')
plt.ylabel('Total Renewable Energy (%)')
plt.grid(True)
plt.show()

# Growth Opportunities Analysis
# Calculate year-over-year growth rates
yearly_growth_rate = total_renewable_energy.pct_change() * 100

# Plot growth rates over time
plt.figure(figsize=(10, 6))
yearly_growth_rate.plot(marker='o', linestyle='-')
plt.title('Renewable Energy Market Growth Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Yearly Growth Rate (%)')
plt.grid(True)
plt.show()

# Key Players Analysis
# Identify top countries with the highest average renewable energy usage
top_countries = data.groupby('Entity')['Renewables (% equivalent primary energy)'].mean().nlargest(5)
print("Top 5 countries with the highest average renewable energy usage:")
print(top_countries)

# Emerging Trends and Technologies Analysis
# Analyze trends in renewable energy usage by country or region
plt.figure(figsize=(12, 8))
sns.boxplot(x='Entity', y='Renewables (% equivalent primary energy)', data=data)
plt.title('Renewable Energy Usage Across Countries')
plt.xlabel('Country')
plt.ylabel('Renewables (%)')
plt.xticks(rotation=45)
plt.show()
```
